```python
import requests
from bs4 import BeautifulSoup

# Setting up the headers for the HTTP request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

# Making a GET request to the URL with headers
response = requests.get('https://www.reliancedigital.in/smartphones/c/S101711?searchQuery=:relevance&page=0', headers=headers)

# Getting the text content of the response
webpage = response.text

# Parsing the HTML content with BeautifulSoup
soup = BeautifulSoup(webpage, 'lxml')

# Now you can use BeautifulSoup methods to parse the HTML
print(soup.prettify())
```

```python
# Empty list to store the key-value pairs
phone_data = []

# Extracting prices from the HTML with class "priceboxw__price"
prices = [i.text.strip() for i in soup.find_all(class_="priceboxw__price")]

# Extracting models from the HTML with class "mss__slider__title"
models = [i.text.strip() for i in soup.find_all(class_="mss__slider__title")]

# Creating key-value pairs (model: price) and storing them in a list of dictionaries
for model, price in zip(models, prices):
    phone_data.append({model: price})

# Returning the list of key-value pairs
df=print(phone_data)
```

```python
# Empty list to store the key-value pairs
phone_data = []

# Extracting prices from the HTML with class "priceboxw__price"
prices = [i.text.strip() for i in soup.find_all(class_="priceboxw__price")]

# Extracting models from the HTML with class "mss__slider__title"
models = [i.text.strip() for i in soup.find_all(class_="mss__slider__title")]

# Creating key-value pairs (model: price) and storing them in a list of dictionaries
for model, price in zip(models, prices):
    phone_data.append({model: price})

# Printing each key-value pair in a new line
for data in phone_data:
    for model, price in data.items():
        print(f"{model}: {price}")
```

```python
import csv

# Empty list to store the key-value pairs
messo = []

# Extracting prices from the HTML with class "priceboxw__price"
prices = [i.text.strip() for i in soup.find_all(class_="priceboxw__price")]

# Extracting models from the HTML with class "mss__slider__title"
models = [i.text.strip() for i in soup.find_all(class_="mss__slider__title")]

# Creating key-value pairs (model: price) and storing them in a list of dictionaries
for model, price in zip(models, prices):
    messo.append({"Model": model, "Price": price})

# Writing the data to CSV format
with open('messo.csv', mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["Model", "Price"])

    # Writing the header
    writer.writeheader()

    # Writing the rows (model and price)
    for data in phone_data:
        writer.writerow(data)

print("Data written to phone_data.csv successfully.")
```
